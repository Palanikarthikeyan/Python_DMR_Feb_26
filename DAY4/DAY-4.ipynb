```python
class box:
    var = 120
    def f2(self):
        print("from f2 method:-->",self.var) # obj.var -> 'data-1'
        self.var = 500 # object based initialization

obj = box()
print(obj.var) # 120
obj.var = 'data-1' # object based initialization
obj.f2() # f2(obj)
print(obj.var) # 500
```
