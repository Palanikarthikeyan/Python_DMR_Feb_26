Write a python program:
i. Create an empty list (ex: products = []) 
ii. use while loop - limit is 5
       iii.  read a product name from <STDIN>
       iv.   append input product to an existing list (step-1)
v.  display list of products and no.of products

```python
products = []
i = 0
while(i < 5):
    p = input('Enter a product name:')
    products.append(p) # list - append operation - adding new data to an existing list
    i=i+1

print('List of products:-')
print(products)
print(f'Total no.of products:{len(products)}')
```

```python
Write a python program:
i. Create an empty dict (ex: products = {}) 
ii. use while loop - limit is 5
       iii.  read a product id  from <STDIN>
       iv.   read a product name from <STDIN>
       iv.   add input product details to an existing dict (step-1) ###  product id - Key  ; product name - Value {'p101':'pA'}
v.  display list of products and no.of products
```

```python
###
# Write a python program:
# read emp.csv file - line by line
# create a newfile - dept.log 
# search - sales dept  # use membership 
# Write matched sales details to dept.log file 
# close the both files


fobj = open('emp.csv','r')
L = fobj.readlines()
fobj.close()

wobj = open('depts.log','w')

for var in L:
    if('sales' in var):
        wobj.write(var) # print(var.strip())

wobj.close()
```
