```python
# List all s3 buckets
import boto3

s3 = boto3.client('s3')
response = s3.list_buckets()

for bucket in response['Buckets']:
    print(bucket)
    #print(bucket['Name'])
```

```python
1. Our python code calls boto3
2. boto3 sends https request to aws api
3. aws authenticates request
4. aws performs action
5. aws sends json response 
6. boto3 converts response into python dictionary 
```

```python
# List EC2 instances
import boto3

ec2 = boto3.client('ec2', region_name='us-east-1')
response = ec2.describe_instances()

for reservation in response['Reservations']:
    for instance in reservation['Instances']:
        print(instance['InstanceId'], instance['State']['Name'])
```
