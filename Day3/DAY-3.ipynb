```python
# Write a python program:
1. initialize a pin Number (ex: pin = 1234)
2. use while loop - limit is 3
       - read a pin Number from <STDIN>
       - test - input pin with an existing pin
       - display - message - pin is matched - count - 123
         Maximum limit is 3
3. if 3 attempt is failed - pin is blocked
|
|
create a newfile pin_history.log file
- append input details to pin_history.log file
- append user input date/time <== time.ctime()
----------------------------------------------------------
```

```python
pin = 1234
count = 0
while(count < 3):
    count = count + 1
    p = input('Enter a pin Number:')
    if(int(p) == pin):
        print(f'PIN - Success - Matched - {count}')
        break # exit from loop

if(pin != int(p)):
    print('Sorry your pin is blocked')
```

```python
fobj = open('demo_pin_history.log','a')

pin = 1234
count = 0
while(count < 3):
    count = count + 1
    p = input('Enter a pin Number:')
    if(int(p) == pin):
        fobj.write(f'PIN Is matched - Entry date /time is:{time.ctime()}\n')
        print(f'PIN - Success - Matched - {count}')
        break # exit from loop
    else:
        fobj.write(f'PIN Is not matched - Entry time is :{time.ctime()}\n')

if(pin != int(p)):
    print('Sorry your pin is blocked')
    fobj.write(f'PIN is not matched')

fobj.close()
```

```python
1. Create new database ->emp.db file
2. Create a table emp - eid,ename,edept,ecity,ecost
3. read - emp.csv file - line by line 
4. split each line into multiple values 
5. insert splitted data into emp Table
6. select and display
-
```

```python
conn = sqlite3.connect('emp.db')
sth = conn.cursor()
sth.execute("create table emp(eid INT,ename TEXT,edept TEXT,ecity TEXT,ecost INT)")
```

```python
for line,data  in enumerate(open('emp.csv','r'),1):
    if(line >1):
        data=data.strip()
        ei,en,ed,ec,ecost = data.split(",")
        sth.execute("insert into emp(eid,ename,edept,ecity,ecost) values(?,?,?,?,?)",(ei,en,ed,ec,ecost))
```

```python
sth.execute("select *from emp")
list(sth)
```

```python
sth.execute("select *from emp where edept = 'sales'")
list(sth)
```

```python
sth.execute("select *from emp where edept = 'sales'")
sales_records = list(sth)
```

```python
wobj = open('sales.log','w')
for var in sales_records: # fetched from DB
    eid,ename,edept,ecity,ecost = var # multiple initialization
    wobj.write(f'{ename}\t{edept}\t{ecity}\t{ecost}\t{int(ecost)*0.8}\n')

wobj.close()
```
